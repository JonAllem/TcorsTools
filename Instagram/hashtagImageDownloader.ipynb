The hashtag scraper needs to be run first and the .pickle files are needed for input.

```python
import pandas as pd
import numpy as np
import tqdm
import urllib.request as req
import pickle
import time
import random
import traceback
import os
from datetime import datetime
```

```python
def fetchImage(url, filePath):
    count = 0
    result = False
    while result is False:
        try:
            req.urlretrieve(url, filePath)
            result = True
        except Exception as e:
            print(e)
            count+=1
            pass
        if count>5:
            print("Unable to retrieve: " + url)
            break
```

```python
hashtagList = ['antibodytest', 'antibodytesting', 'covidtest', 'covidtesting'];
```

```python
with open('data_antibodytest.pickle', 'rb') as file:
    data_antibodytest = pickle.load(file)
with open('data_antibodytesting.pickle', 'rb') as file:
    data_antibodytesting = pickle.load(file)
with open('data_covidtest.pickle', 'rb') as file:
    data_covidtest = pickle.load(file)
with open('data_covidtesting.pickle', 'rb') as file:
    data_covidtesting = pickle.load(file)
dataList = [data_antibodytest, data_antibodytesting, data_covidtest, data_covidtesting]
```

```python
itr = 3
```

```python
data = dataList[itr]
hashtag = hashtagList[itr]
```

```python
urls = []
vid_count = 0
img_count = 0
slide_count = 0
for post in data:
    if post['postType'] == 'GraphVideo':
        vid_count+=1
        urls.append('-' + post['videoURL'])
    elif post['postType'] == 'GraphSidecar':
        slide_count += 1
        urls.append(post['imgURLList'])
    elif post['postType'] == 'GraphImage':
        img_count+=1
        urls.append(post['imageLink'])
```

```python
vid_count, img_count, slide_count
```

```python
curPath = 'image/' + hashtag + '/'
print("Current filepath: " + curPath)
# os.mkdir(curPath)
```

```python
for j in tqdm.trange(4199,len(urls[0:])):
    post = urls[j]
    if type(post) == str:
        if post[0] == '-':
            fetchImage(post[1:], curPath + str(j+1) + '.mp4')
        else:
            fetchImage(post, curPath + str(j+1) + '.jpg')
    else:
        os.mkdir(curPath + str(j+1) + '/')
        for k in range(len(post)):
            url = post[k]
            fetchImage(url, curPath + str(j+1) + '/' + str(k+1) + '.jpg')
    if (j+1)%50 == 0:
        time.sleep(random.uniform(1,2))
```

```python
tempDf = pd.DataFrame(data)
tempDf['accessibility_caption'] = tempDf['caption']
tempDf = tempDf.drop(['imgURLList', 'videoURL', 'videoViewCount', 'caption', 'imageLink'], axis=1)
tempDf['timeStamp'] = pd.to_datetime(tempDf['timeStamp'], unit='s')
tempDf.to_excel(hashtag + '_posts.xlsx', index=False)
```
